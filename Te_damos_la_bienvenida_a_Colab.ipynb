```
import numpy as np
import matplotlib.pyplot as plt

# Parámetros del problema
k = 0.4        # constante de crecimiento
P0 = 100       # población inicial
t0 = 0         # tiempo inicial
tf = 5         # tiempo final
h = 0.1        # tamaño de paso para Euler

# Definición de la función f(t, P)
def f(t, P):
    return k * P

# Solución analítica
def P_analitica(t):
    return P0 * np.exp(k * t)

# Método de Euler
def euler(f, t0, tf, P0, h):
    t_values = np.arange(t0, tf + h, h)
    P_values = np.zeros(len(t_values))
    P_values[0] = P0

    for i in range(1, len(t_values)):
        P_values[i] = P_values[i-1] + h * f(t_values[i-1], P_values[i-1])

    return t_values, P_values

# Ejecutar método de Euler
t_euler, P_euler = euler(f, t0, tf, P0, h)

# Crear vector de tiempo para solución analítica
t_analitica = np.linspace(t0, tf, 100)
P_exacta = P_analitica(t_analitica)

# Graficar ambas soluciones
plt.figure(figsize=(10,6))
plt.plot(t_analitica, P_exacta, label="Solución Analítica", color='blue')
plt.plot(t_euler, P_euler, 'o--', label="Método de Euler (h=0.1)", color='red')
plt.xlabel('Tiempo (horas)')
plt.ylabel('Población bacteriana')
plt.title('Crecimiento de la población bacteriana')
plt.legend()
plt.grid()
plt.show()

# Imprimir algunos resultados
for i in range(len(t_euler)):
    print(f"t = {t_euler[i]:.1f} h, P (Euler) = {P_euler[i]:.2f}")
```
